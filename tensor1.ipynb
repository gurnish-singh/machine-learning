```

import tensorflow as tf
from tensorflow import keras
```

```
print(tf.__version__)
```

```
model=keras.Sequential([keras.layers.Dense(units=1,input_shape=[1])])
#units =1 means there is only one neuron
#keras is the library to define a neural network
##here we r defining layers
#input_shape gives us the shape of the input layer obviously.
```

```
model.compile(optimizer='sgd',loss='mean_squared_error')
#here is the math part.
#here we r defing the loss function (cost for the guess :) )
# optimizer picks up the next guess
#here sgd is the stochastic gradient descent
```

```
import numpy as np
x=np.array([-1,0,1,2,3,4],dtype=float)
y=np.array([-3,-1,1,3,5,7],dtype=float)
#so here is our data
```

```
# now we do training 
model.fit(x,y,epochs=500)
#it will train for 500 times
```

```
# now lets test our model
print(model.predict([10]))
```

answer came out to be 18.97 because we trained it on very little data :)
